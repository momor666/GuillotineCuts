```python
from BP2FP import BP2FP
from separability import isSeparable
permstr = "4 14 9 13 15 10 11 5 1 3 8 12 6 2 7"
print("Is separable:", isSeparable(permstr))
n = float(len(permstr.split()))
rects = BP2FP(permstr)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig = plt.figure(figsize=(10,10))
ax = plt.Axes(fig, [0, 0, 1, 1])
fig.add_axes(ax)
for key in rects:
    rect = tuple(v / n for v in rects[key])
    w = rect[1] - rect[0]
    h = rect[3] - rect[2]
    ax.add_patch(patches.Rectangle((rect[0], rect[2]), w, h, fc='none', ec='black'))
    
    cx = rect[0] + w / 2.0
    cy = rect[2] + h / 2.0
    ax.annotate(key, (cx, cy), color='b', fontsize=15, ha='center', va='center')
```
